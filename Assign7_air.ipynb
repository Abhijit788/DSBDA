```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/AirQuality_visualization.csv',delimiter=';')
```

```python
df
```

```python
df = df.rename(columns={'T': 'Temperature'})
df = df.rename(columns={'RH': 'Relative Humidity'})
df = df.rename(columns={'AH': 'Absolute Humidity'})
df
```

```python
df = df.drop(['Unnamed: 15', 'Unnamed: 16'],axis=1)
```

```python
df
```

```python
df['CO(GT)'] = df['CO(GT)'].str.replace(',', '.').astype(float)
df['C6H6(GT)'] = df['C6H6(GT)'].str.replace(',', '.').astype(float)
df['Temperature'] = df['Temperature'].str.replace(',', '.').astype(float)
df['Relative Humidity'] = df['Relative Humidity'].str.replace(',', '.').astype(float)
df['Absolute Humidity'] = df['Absolute Humidity'].str.replace(',', '.').astype(float)
df
```

```python
df.info()
```

```python
df.head()
```

```python
df=df.drop_duplicates()
```

```python
df.isna().sum()
```

```python
# Convert columns containing numeric data to numeric types explicitly
for col in ['CO(GT)', 'C6H6(GT)', 'Temperature', 'Relative Humidity', 'Absolute Humidity', 'PT08.S4(NO2)', 'PT08.S5(O3)', 'PT08.S2(NMHC)', 'PT08.S1(CO)', 'NOx(GT)', 'NO2(GT)']:
    df[col] = pd.to_numeric(df[col], errors='coerce')
# Now fillna with the mean, this will only consider numeric columns
# Convert 'Date' and 'Time' columns to datetime objects
df['DateTime'] = pd.to_datetime(df['Date'] + ' ' + df['Time'], format='%d/%m/%Y %H.%M.%S')
# Set 'DateTime' as index
df = df.set_index('DateTime')

# Calculate the mean for numeric columns only
numeric_cols = df.select_dtypes(include=np.number).columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].mean())

# Drop remaining rows with any missing values
df = df.dropna()
```

```python
df.isna().sum()
```

```python
df
```

```python
df['Absolute Humidity'] = df['Absolute Humidity'].multiply(100)
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def remove_outliers(column):
    Q1 = column.quantile(0.25)
    Q3 = column.quantile(0.75)
    IQR = Q3 - Q1
    threshold = 1.5 * IQR
    outlier_mask = (column < Q1 - threshold) | (column > Q3 + threshold)
    return column[~outlier_mask]
```

```python
df.columns
```

```python
# Remove outliers for each column using a loop
col_name = ['Temperature','Relative Humidity','Absolute Humidity','PT08.S4(NO2)','PT08.S5(O3)','C6H6(GT)',
       'PT08.S2(NMHC)','PT08.S1(CO)']
for col in col_name:
    df[col] = remove_outliers(df[col])
```

```python
df['Year'] = pd.to_datetime(df['Date'], format='%d/%m/%Y').dt.year
df['Month'] = pd.to_datetime(df['Date'], format='%d/%m/%Y').dt.month

df
```

```python
df['yearr']= df.Year.astype(str)
df['month']= df.Month.astype(str)
```

```python
# Plot correlation heatmap
plt.figure(figsize=(8, 6))
# Select only numeric columns for correlation calculation
numeric_df = df.select_dtypes(include=np.number)
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
